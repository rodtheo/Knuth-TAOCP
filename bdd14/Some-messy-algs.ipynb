```python
# ALGORITHM B - Solutions of maximum weight (7.1.4-pg209)

I = {0: (5, 0, 0), 1: (5, 1, 1), 2: (4, 0, 1), 3: (4, 1, 0), 4: (3, 3, 2),
     5: (3, 1, 0), 6: (2, 0, 1), 7: (2, 5, 4), 8: (1, 7, 6)}

# vector w from w_{0},...,w_{n}
w = [0, 1, -2, -3, 4]

n = 4
s = len(I)
# Initializing auxiliary vectors
m = range(s)
W = range(n+2)
t = range(s)
# Step B1
# W_{n+1} = 0
# Como o python eh 0-based subtraimos 1 do indice
W[n+1] = 0
# W_j = W_{j+1} + max(w_{j}, 0) para todo n >= j >= 1
for j in reversed(xrange(1,n+1)):
    print j
    W[j] = W[j+1] + max(w[j], 0)

# STEP B2 [loop over k;]  
# set m_{1}=0
m[0] = 0
m[1] = 0

# STEP B3 [process I_{k}]
print("\tv\tl\th\tt_k\tm_k\tm\n")
for k in xrange(2, s):
    v, l, h = I[k]
    t[k] = 0
    if l != 0:
        m[k] = m[l] + W[v+1] - W[I[l][0]]
    if h != 0:
        m_v = m[h] + W[v+1] - W[I[h][0]] + w[v]
        if (l==0) or (m_v>m[k]):
            m[k] = m_v
            t[k] = 1
    print("K={}\t{}\t{}\t{}\tt_{}={}\tm_{}={}\t{}".format(k, v, l, h, k, t[k], k, m[k], m_v))
# STEP B4
j = 0
k = s-1
x=range(n+1)
while (j != n):
    while (j < I[k][0]-1):
        j = j + 1
        if w[j] > 0:
            x[j] = 1
        else:
            x[j] = 0
    if k > 1:
        j = j+1
        x[j] = t[k]
        if t[k] == 0:
            k = I[k][1]
        else:
            k = I[k][2]
print("x=",x[1:])
```
